# Data Cleaning

## Objective

The objective of this notebook is to prepare the integrated adaptations dataset for exploratory data analysis by removing invalid records, correcting inconsistencies, and standardizing the dataset.

## Cleaning Steps

* Removed records marked for exclusion during manual verification.
* Replaced incorrect book titles with manually corrected values.
* Removed unnecessary columns used during the data collection process.
* Converted release dates to datetime format.
* Verified missing values and duplicate records.
* Created the final cleaned dataset for analysis.


```python
import pandas as pd
```

```python
df = pd.read_csv('../data/books/final/updated_adaptations.csv')

df.info()
```

```python
df.head()
```

## Removing Non-Adaptation Titles

```python
df.value_counts('keep')
```

```python
df = df[df['keep'].str.lower() == 'y']
```

## Correcting Book Titles

```python
df = df.drop(columns=['book_title'])
```

```python
df = df.rename(columns={'search_key': 'book_title'})
```

## Dropping Unnecessary Columns

```python
df = df.drop(columns=['is_book_adaptation','keep'])
```

```python
df = df.reset_index(drop=True)
```

```python
df.info()
```

## Converting Data Types

```python
df['Premiere'] = pd.to_datetime(df['Premiere'], format='%d/%m/%Y')
```

## Final Dataset Validation

```python
df = df [
    [
        "Title",
        "Premiere",
        "Rated",
        "imdb_rating",
        "imdb_votes",
        "imdb_genre",
        "imdb_year",
        "imdb_type",
        "platform",
        "book_title",
        "book_author",
        "goodreads_rating"
    ]
]
```

```python
df = df.reset_index(drop=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df.to_csv('../data/final/book_adaptations.csv', index=False)
```

## Final Dataset Summary

The final cleaned dataset contains 248 verified book adaptations from Netflix and Amazon Prime Video, enriched with IMDb and Goodreads ratings and prepared for exploratory data analysis.
